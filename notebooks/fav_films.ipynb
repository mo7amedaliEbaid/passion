# My Favorite Films

```python
import pandas as pd

# Load the CSV file
films = pd.read_csv('favorite_films.csv')

# Display the first few rows of the dataframe
films.head()
```

```python
import matplotlib.pyplot as plt

# Count the number of films by each director
director_counts = films['Directors'].value_counts()

# Create a bar plot for the number of films by each director
plt.figure(figsize=(12, 6))
director_counts.plot(kind='bar', color='skyblue')
plt.xlabel('Directors')
plt.ylabel('Number of Films')
plt.title('Number of Films by Each Director')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
# Count the number of films released in each year
year_counts = films['Year'].value_counts().sort_index()

# Create a bar plot for the number of films released in each year
plt.figure(figsize=(12, 6))
year_counts.plot(kind='bar', color='lightgreen')
plt.xlabel('Year')
plt.ylabel('Number of Films')
plt.title('Number of Films Released Each Year')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from collections import Counter

# Count the number of films in each genre
genre_counts = Counter()
for genres in films['Genres']:
    for genre in genres.split(', '):
        genre_counts[genre] += 1

# Create a bar plot for the number of films in each genre
plt.figure(figsize=(12, 6))
plt.bar(genre_counts.keys(), genre_counts.values(), color='salmon')
plt.xlabel('Genre')
plt.ylabel('Number of Films')
plt.title('Number of Films by Genre')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```
