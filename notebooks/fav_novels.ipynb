# My Favorite Books

## 1. One Hundred Years of Solitude
- **Author**: Gabriel García Márquez
- **Year of Publish**: 1967
- **Overview**: This novel tells the multi-generational story of the Buendía family, whose patriarch, José Arcadio Buendía, founded the town of Macondo in the middle of a jungle in Colombia. The story is a rich tapestry of magical realism, exploring themes of time, fate, and the nature of reality.
- **Image**:  
![One Hundred Years of Solitude](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcS8-1DdgXVkNImBgNEIg_amH6uJB5teasLAG5zdCe5LvTTSLibTLR4g52BzB_xFGz6iwwI&usqp=CAU)

## 2. 1984
- **Author**: George Orwell
- **Year of Publish**: 1949
- **Overview**: A dystopian social science fiction novel and cautionary tale about the future, 1984 describes a totalitarian regime that employs surveillance, censorship, and repression to exert complete control over the citizens of Oceania. The story follows Winston Smith as he navigates this oppressive world.
- **Image**:  
![1984](https://images.penguinrandomhouse.com/cover/9780451524935)

## 3. The Metamorphosis
- **Author**: Franz Kafka
- **Year of Publish**: 1915
- **Overview**: The Metamorphosis is a novella that follows the unsettling transformation of Gregor Samsa, a traveling salesman, who wakes up one morning to find himself transformed into a giant insect. The story explores themes of alienation, identity, and the human condition.
- **Image**:  
![The Metamorphosis](https://images.penguinrandomhouse.com/cover/9780143105244)

## 4. The Brothers Karamazov
- **Author**: Fyodor Dostoevsky
- **Year of Publish**: 1880
- **Overview**: This philosophical novel delves into the complex dynamics of the Karamazov family. It explores profound themes such as faith, doubt, free will, and the nature of good and evil. The story revolves around the murder of the father, Fyodor Karamazov, and the moral and existential struggles of his sons.
- **Image**:  
![The Brothers Karamazov](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcS6sOKAtHh_BrEO3wODCW0PQJN15e57caLBdm8BhK9t3PTyWRG9v5xCkP3K4kQv7GL8mQw&usqp=CAU)

## 5. The Stranger
- **Author**: Albert Camus
- **Year of Publish**: 1942
- **Overview**: The Stranger is a novel that explores the themes of absurdism and existentialism. It follows the story of Meursault, an indifferent French Algerian who is emotionally detached from the world around him. The novel begins with the death of his mother and ends with Meursault's own trial for murder.
- **Image**:  
![The Stranger](https://images.penguinrandomhouse.com/cover/9780679720201)

## 6. War and Peace
- **Author**: Leo Tolstoy
- **Year of Publish**: 1869
- **Overview**: War and Peace is a novel that intertwines the lives of several families during the Napoleonic Wars in Russia. It explores themes of history, destiny, and the human condition, and is renowned for its complex characters and detailed depiction of early 19th-century Russian society.
- **Image**:  
![War and Peace](https://encrypted-tbn0.gstatic.com/images?q=tbn:ANd9GcRPVrT7OceepOkHLr0TyhJI3b3tDjqHgyYgPy7Oj35nXJEdU-omphV6bsZJx7EQrfVBhAo&usqp=CAU)

## 7. Crime and Punishment
- **Author**: Fyodor Dostoevsky
- **Year of Publish**: 1866
- **Overview**: This psychological drama focuses on the mental anguish and moral dilemmas of Rodion Raskolnikov, an impoverished ex-student in Saint Petersburg who formulates and executes a plan to kill an unscrupulous pawnbroker for her money.
- **Image**:  
![Crime and Punishment](https://upload.wikimedia.org/wikipedia/en/4/4b/Crimeandpunishmentcover.png)

## 8. The Color Purple
- **Author**: Alice Walker
- **Year of Publish**: 1982
- **Overview**: The Color Purple is an epistolary novel that chronicles the struggles of African American women in the early 20th century American South. The story is told through the letters of Celie, who endures abuse and oppression but ultimately finds her voice and identity.
- **Image**:  
![The Color Purple](https://upload.wikimedia.org/wikipedia/en/thumb/6/60/ColorPurple.jpg/220px-ColorPurple.jpg)

## 9. Animal Farm
- **Author**: George Orwell
- **Year of Publish**: 1945
- **Overview**: Animal Farm is an allegorical novella that uses a group of farm animals who overthrow their human farmer in order to create a society where the animals can be equal, free, and happy. The story is a critique of totalitarianism and corruption.
- **Image**:  
![Animal Farm](https://trey-stone.com/wp-content/uploads/2022/02/3149348.jpg)

## 10. Brave New World
- **Author**: Aldous Huxley
- **Year of Publish**: 1932
- **Overview**: Brave New World is a dystopian novel set in a future world state where citizens are environmentally engineered into an intelligence-based social hierarchy, and the story explores the consequences of a technologically advanced, socially stratified society.
- **Image**:  
![Brave New World](https://upload.wikimedia.org/wikipedia/en/6/62/BraveNewWorld_FirstEdition.jpg)


```python
import matplotlib.pyplot as plt

# Data for the authors and the number of books
authors = ["Gabriel García Márquez", "George Orwell", "Franz Kafka", "Fyodor Dostoevsky", "Albert Camus", "Leo Tolstoy", "Alice Walker", "Aldous Huxley"]
num_books = [1, 2, 1, 2, 1, 1, 1, 1]

# Create the bar plot
plt.figure(figsize=(10, 6))
plt.bar(authors, num_books, color='skyblue')
plt.xlabel('Authors')
plt.ylabel('Number of Books')
plt.title('Number of Books by Each Author in the List')
plt.xticks(rotation=45, ha='right')
plt.grid(axis='y')

# Show the plot
plt.tight_layout()
plt.show()
```
